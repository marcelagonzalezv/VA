## Tarea 2 - Preprocesamiento de los datos

```python
# import libraries
import pandas as pd
import pandas_profiling
import numpy as np
```

```python
FileName = "Resultados_Saber_Pro_Competencias_Genericas_2018-2.csv"
cols=[
'ESTU_AREARESIDE',
'ESTU_CONSECUTIVO',
'ESTU_GENERO',
'ESTU_INST_DEPARTAMENTO',
'ESTU_METODO_PRGM',
'ESTU_NACIONALIDAD',
'ESTU_VALORMATRICULAUNIVERSIDAD',
'MOD_INGLES_DESEM',
'MOD_INGLES_PUNT',
'INST_COD_INSTITUCION',
'INST_NOMBRE_INSTITUCION',
'PERIODO'
]
df2018 = pd.read_csv(FileName, usecols= cols )
df2018.head()
```

```python
# GROUP BY 
t1 = df2018.groupby(['ESTU_INST_DEPARTAMENTO', 'ESTU_METODO_PRGM']).agg({'MOD_INGLES_PUNT': 'max'})
t1.to_csv('modalidad_max.csv')
t1.head(10)
```

```python
t2 = df2018.groupby(['ESTU_INST_DEPARTAMENTO']).agg({'MOD_INGLES_PUNT': 'mean'})
t2.to_csv('depto_mean.csv')
t2.head(10)
```
